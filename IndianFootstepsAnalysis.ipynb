### 6)Hypothesis testing using student's t test

let,

Hypothesis A: Population of India walks an average of 10,000 footsteps everyday.

Hypothesis B: Population of India Does NOT walk an average of 10,000 footsteps everyday.

```python
mu=len(df['Footsteps'])-1 #the degrees of freedom --> mu
hypo_mean=10000
sl=0.05  #significance level/confidence

print('The Standard Deviation for the given sample is :',std_dev_df1)
print('The mean for the given sample is:',mean_df1)
print('The significance level for this test is:',sl)
print('The Degrees of Freedom for this sample is:',mu)
print('The average footsteps need to be taken is:',hypo_mean)
```

```python
def compute_t(a,b,c,d):
    e=(a-b)/(c/d**1/2)
    return e
t_value=compute_t(mean_df1,hypo_mean,std_dev_df1,mu)
print('The t-values for the sample is:',t_value)
```

```python
t_critical=scipy.stats.t.ppf(q=sl,df=mu)
print('The value of t critical is:',t_critical)
```

Since the t value of the sample is less than the t critical value , we can reject the hypothesis A with a 95% confidence.
